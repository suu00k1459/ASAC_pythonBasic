```python
import pandas as pd
import numpy as np
```

```python
# 1. scikit -learn에서 필요한 모듈을 불러다 사용하면 됨
# 1-1 샘플 데이터 : 자주 변경됨
from sklearn.datasets import fetch_openml


# 1-2 주어진 데이터 분리 : train_test
from sklearn.model_selection import train_test_split


# 1-3 내가 사용할 모델 f : knn 모델을 불러와서
# 참고) 파이썬 or scikit-learn 계열 : 분류 -> classifier, regressor

from sklearn.neighbors import KNeighborsClassifier

# 1-4 평가 : 여러가지 평가 지표 -> 간단하게 accuracy etc
from sklearn.metrics import accuracy_score
```

```python
mnist = fetch_openml( "mnist_784",version=1, as_frame=False,
                     parser="auto")
mnist
```

```python
X=mnist["data"]
X.shape
```

```python
# ML을 하기 위해서는 1개 샘플은 무조건 1D 차원 벡터로 변환
# 여러 샘플이 모인 것은 2D의 행렬로 처리가 되어야 함.

y = mnist["target"]
y.shape
```

```python
# 주어진 데이터 셋을 f를 찾는데 사용할지, 찾은 f를 평가하는데 사용할지
# 모의고사 문제, 실전 문제, 모의고사 정답
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.25, #전체 x 중에서 25르르 test로 해줌
                                                   random_state= 1234)
```

```python
# 1개의 샘플이 2차원 데이터였는데 ml을 하기 이해서 1차로 변호나
# EDA 를 위해서는 2D로 다시 역변환 해야함

import seaborn as sns

sns.heatmap(train_X[1].reshape(28,28))
```

```python
train_y[1]
```

```python
# 실제로는 : EDA를 타이트하게 진행해야 함
# 새로운 변수들도 생성을 하면서 진행할 것.
```

```python
# 정답 : 정답들에 대한 분포
# pandas에서 할 때 ; values_counts()
# 정답의 종류는 10개 : 다중클래스 분류
# 각각의 비율  :대략 10개가 비슷, 한 쪽을 쏠린 것
# - 한쪽으로 쏠린 경우  : imbalance problem
    # 평가지표 : accuracy를 써도 괜찮겠따... 성능은 낮을수 있겠다

np.unique(train_y, return_counts = True)
```

```python
# 정답에 대한 비율, 분포를 체크해야함
# -> 모델링 하기 전 처리할 작업 or 지표들에 대한 고민
```

```python
# 돌리려는 모델을 세팅
# f -- > knn 분류 : KNeighborsClassifier
# 사용할 모델을 불러오기
knn  = KNeighborsClassifier(n_neighbors=5, n_jobs=-1) # n_jobs =-1 최대코어 사용함

# 참고) 구체적인 f에 대한 세팅을 해서 주어진 데이터에 대한 f알고리즘의 최적의 성능을 뽑아내기 위한 과정
    #  hyper parameter tunning
    # 직접 하나씩 탐색: gridsearch
    # 랜덤 조합 탐색 : randomGRidsearch
    # + bayesian opt
    # AutoML( 피쳐 처리!)
```

```python
# 2) 학습을 수행 : 주어진 데이터에 맞춰서
#                  내가 세팅한(디테일하게 세팅한) f에 대한
#                  파라미터를 최적화 : Learning
# ==> 시간이 오래 걸리는 부분...
# 학습할 문제와 학습한 문제와 순서가 일치한 정답지
knn.fit(train_X, train_y)
```

```python
# 3) 2번에서 공부한 것을 바탕으로 실전문제를 풀어
#    ==> predict
# 수능 시험장의 OMR카드
pred = knn.predict(test_X)
```

```python
accuracy_score(test_y, pred)
```

```python
# 대충 knn 이라는 모델을 바탕으로 데이터에 적용해보자
# baseline =0.97

# 0.97보다 더 좋은 성능이 나오도록 knn의 디테일한 파라미터들을 바꿔서 찾아보자
# 경우에 따라 한 번 학습하는 게 아니라  k fold cv 제거

# k=fold CV
from sklearn.model_selection import cross_val_score
```

```python
# 직접 하나씩 구현을 해보면
# 1) 테스트할 파라미터의 값들을 리스트 업
neighbors = [ i for i in range(1,10) if i %2==1]

# 2) 1개 파라미터 갑으로 했을 때의 결과들을 모을 변수
cv_scores = []

# 3) 테스트할 값들을 돌리면 됨
for k in neighbors:
    # 3-1) 모델을 준비
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)

    # 3-2) k-fold를 이용해서 train 데이터를 k 번 학습, 평가
    # -> 어떤 모델, 문제, 정답, 몇 등분할지(HP)
    # 참고) 채점 기준 일반적인 : 문자열, 새로운 것은 함수
    
    scores = cross_val_score(knn, train_X, train_y, cv=10, scoring = "accuracy")

    # scores 에는 k=1에 knn에 대해서 10번에 대한 acc
    cv_scores.append(scores.mean())
    print(str(k)+"k values Accuracy:", scores.mea())

```

```python
# 시도할 값들에 대한 리스트업과 그에 따른 반복문
# gridsearchCV, 하나의 함수로 처리를 함!

# KNN : 거의 사용 안 함!!!
# SVM : 거의 잘 사용을 안 함!!!!
# DT  : 잘 사용을 안 함!!!!!!
# -----------------------------------------------------
# 여러 모델을 활용한 모델
# RandomForest : 여러 모델을 활용!!!(Bagging)
# xgboost , lightGBM, catBoost : 여러 모델을 활용(Boosting)
# ---> 현실적으로 주로 사용되는 모델들!!!
# 공통점 : Tree기반으로 하는 모델.....
#--------------------------------------------------------
# ==> 최종 모델을 선정!!!!!!
# 1) 했던 것들 중에서 best 1개 선정!!!!
# 2) 했던 여러 모델을 종합해서
#     voting : Hard/Soft
#     Stacking : 대회에서만 종종 쓴느 경향..(현업에는 굳이..)
# +++ Bagging 것들을 스케일을 업해서 하는 방법.....
# ==> 여러 모델을 활용해서 Genenral!!!!!!

# +++ 개별 모델에 대한 성능 극대화 : HPT
```
