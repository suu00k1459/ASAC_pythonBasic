```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

from selenium.webdriver.common.by import By

# 키보드 조작이 필요하면, 키보드 입력을 할 수 있는
from selenium.webdriver.common.keys import Keys
```

```python
# webdriver 경로 지정
path = r"C:\Users\NT551XCJ\Desktop\chromedriver-win64\chromedriver.exe"
```

```python
# 브라우저를 통해서 컨트롤 하기 위한 작업
s = Service(path)
driver = webdriver.Chrome( service = s)

# 브라우저에서 할 수 있는 여러가지 일들
# -> 간단하게 창 크기 조절
driver.set_window_size(300,300)

# 내가 접속하려는 사이트
url = "https://naver.com"
driver.get(url)
```

```python
driver.page_source #연결이 되어 있어야 볼 수 있음.
```

```python
# -> 검색창에 검색어를 넣고 싶은 것!
element = driver.find_element(By.XPATH, '//*[@id="query"]' )  #우클릭 Copy - Copy Xpath
element.send_keys("bts")

# -> 검색을 해달라는 클릭을 찾아서 클릭해줘!
driver.find_element(By.XPATH, '//*[@id="search-btn"]').click()
```

```python
# 종료를 하기 위해서
driver.quit()
```
