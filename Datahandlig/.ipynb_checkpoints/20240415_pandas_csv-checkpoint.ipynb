```python
# Q) 01_olive 파일을 불러와 보세요. df 로...
file3 = path + "\\01_olive.csv"
data = pd.read_csv(file3, sep = ",")
data.head()
```

```python
# 일단 df로 받아왔는데, 처음 컬럼에 값이 없어서 unnamed :0 으로 컬럼 이름으로 들어옴
# -> 기존의 컬럼을 변경하기
# unnamed P:0 -> id_area
# region -> code

data.rename(columns  = {"Unnamed: 0" : "id_area", "region" : "code"}, inplace=True)
```

```python
data.tail(3)
```

```python
# id_area 컬럼이 중복된 값이 있는지 체크

len(data.loc[:,"id_area"].unique()) == len(data)
```

```python
# Q) 간단한 전처리
# id_area 컬럼의 값들을 파있너을 기준으로 기준 숫자들을 변경
# 1.north_apulia -> 0.north_apulia
# 2.north_apulia -> 1.north_apulia
# ...
# 572.West-Liguria -> 571.West-Liguria
# ->그렇다고 for문? no!
# -> apply + lambda 같은 것으로 구현할 수 있음!
# 규칙이 존재하기 때문!
# 규칙 : 앞의 숫자를 -1로 줄이고, 뒤의 지역은 그대로!


data["id_area"] = data["id_area"].apply(lambda x : str(int(x.split(".")[0])-1)+"."+x.split(".")[1])
```

```python
data
```

```python
# Q) palmitic 컬럼의 값을 지금은 백~천 단위의 값이 되어 있음.
#    --> 단위 환산을 통해서 /100으로 범위 재조정!!!!
#    1075 --> 10.75
#    1088 --> 10.88
#     911 ---> 9.11
# ==> 단위 환상& 변경

data.loc[:, "palmitic"]/100 #벡터연산 가능
```

```python
# Q) id_area 컬럼이 유니크하니깐, 가로줄에 index로 지정!
# -> 내가 만든 가로줄 인덱스로 접근하기 위해서!
# 기존 컬럼을 가로줄 인덱스로 역할 변경!
# (이미 유니크한지는 체크를 하였음)
# set_index

data.set_index("id_area", inplace=True)
```

```python
# 내가 할 전처리나 뭔가 작업을 다 했다고 하고,,
# --> 다른 사람한테 전달, 다른 데이터 작업을 하고 나중에 처리하는 등의 경우

data.to_csv(path+"\\..\\test_csv.csv")
# 참고) 한글이 있는 df인 경우에는 출력할 때 출력하려는 os나 사용할 프로그램을 보고 encoding 에 대한 방식도 추가 옵션으로 하면 됨

# 기타 옵션
```
