```python
import pymysql
import pandas as pd
```

```python
# mysql 이 서버가 내부적으로 동작하고 있어야 함.
# 작업관리자에서 확인해보기
# 접속 관련 연결설정!

mysql_con = pymysql.connect(
    user = "root",
    passwd = "0000",
    host = "127.0.0.1", #접속하려고 하는 mysql 의 ip주소: 내 pc
    db = "sqldb",
    charset = "utf8" # 데이터에 한글이 있을 때, 파이썬 쪽에서 안 깨지려고 사용,,
)
```

```python
# 연결
my_cursor = mysql_con.cursor(pymysql.cursors.DictCursor)
```

```python
# 내가 요청할 쿼리문 작성
sql = "select * from usertbl"
```

```python
my_cursor.execute(sql)
```

```python
# 결과 중에서 내가 가지고 올 데이터들
result = my_cursor.fetchall()
result
```

```python
df = pd.DataFrame(result)
```

```python
df
```

```python
# --> mysql 에서 사용하는 필드의 속성과는 100퍼센트 동일하게 오지는 않음
# 적당히 pandas에서 사용하는 타입들로 전환
# 데이터 변환이 제대로 오는지 확인
```

```python
# 할 일 끝나고 접속 끊을 때
my_cursor.close()
```

```python
# 참고 : pandas 에서 직접 sql을 요청할 수도 있음..
# -> QLAlchemy 패키지를 써서 접속하고, 요청을 해야 함...
# 저는 pymysql 을 선호(취향.)
```

```python
# 경고가 뜨니 자료형에 유의해서 사용!!!
pd.read_sql_query(sql, mysql_con)
```
