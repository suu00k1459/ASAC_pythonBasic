```python
# 목적1
# 앞에서 kobis api 에서 영화목록에 대한 요청을 하고 df로 처리하는 과정에 있어서
# + 일단 있는 정보는 다 가지고 와서 처리하자!

# 목적2
# urllib 대신 requests  사용하기
# -> json 에서는 이미 내부에 자동으로 변환하는 모듈이 있음!
# import json은 생략이 가능함
```

```python
# 필요한 모듈 : urllib 의 모듈이랑 좀 다름

import pandas as pd
import requests
# import json
```

```python
# requests 패키지로 접속하는 방법 중 하나
# 1) 메뉴얼상에 내가 필요한 api에 base_url

base_url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
# 2) 실제 요청하는 파라미터들에 대한 요청 사항을 dict로 표현!!!!
#    api 메뉴얼상 요청인터페이스 인자 : key(문자열)
                # 요청인자에 해당하는 요청 값 : values(숫자, 문자)

my_params = {
    "key" :"68195af132b8cc46b7dc02955585c26f",
    "itemPerPage":40
}

# 3) 실제 요청을 해야 함!
res = requests.get(base_url, params = my_params)

# 4) FM으로는 http을 요청하면, 응답 코드 중 200(ok) 체크하기
if res.ok:
    result = res.json()
else:
    print("check your http response error!")

result
```

```python
# --> 전체 영화 정보들을 담을 리스트
new_list=[]

# idx = 정수 인덱스 : 감독 정보를 좀 처리하려고...
# data = 개별 영화에 대한 정보 : dict {}
for idx, data in enumerate(result["movieListResult"]["movieList"]):
    one_data = data
    new_list.append(one_data)
print(new_list)
```

```python
pd.DataFrame(new_list)
```

```python
# ++ dict 를 활용해서 처리할 때
# => 감독에 대한 1번 감독이름만 처리하자!

directors_list =[]
for idx, data in enumerate(result["movieListResult"]["movieList"]):
    directors_list.append(data["directors"])
    
```

```python
# 참고)
# dict의 병합
# -> key 가 중복이면, 값을 갱신, 중복이 아니면 추가됨!!!!!!!!!

# * 기호 본 경험
# 1) *args  : 파이썬 sum_random(*args): 리스트
# 2) **args : def aaa(p="12", cc=123)  **args
```

```python
for idx, data in enumerate(result["movieListResult"]["movieList"]):
    one_data = data
    if len(directors_list[idx]) != 0:
        one_data = {**one_data, **directors_list[idx][0]}
        new_list.append(one_data)

pd.DataFrame(new_list)
```

```python
# 일단 위의 코드들은 기본적으로 수집한 데이터를 다 df로 담는
#      ->정말로 안 쓸 컬럼들은 지울 수 있음!
# df에서 컬럼을 지우는 것 : drop axis = 0/1

pd.DataFrame(new_list).drop(["directors", "companys"], axis =1).fillna("")
# ==> DF 이면 계속 원하는 변경들에 대한 메서들 적용할 수 있음.
```
